# Algoritmos

Escribe un algoritmo y realiza un diagrama de flujos para calcular el promedio de las calificaciones de la materia de programación

***ALGORITMO***

1- El usuario ingresará cada una de sus calificaciones del semestre de programación. (para cada input) (serán 4 calificaciones para este caso)

2- La maquina sumará todas las calificaciones y las divirá entre 4. Esto es, el promedio de las calificaciones.

3- mostraremos el promedio resultado.

````python
# El usuario ingresa cada calificacion en su parcial
C1 = float(input("Ingrese la primera calificación: "))
C2 = float(input("Ingrese la segunda calificación: "))
C3 = float(input("Ingrese la tercera calificación: "))
C4 = float(input("Ingrese la cuarta calificación: "))

# Se calcula el promedio de las calificaciones
promedio = (C1 + C2 + C3 + C4) / 4

# Se muestra el resultado
print("El promedio de las calificaciones es:", promedio)

# EXTRA. Una ayudita al alumno :)
print("Su calificación final:", round(promedio))
````

* **Define un método que tenga como meta para el usuario adivinar el numero generado aleatoriamente entre 1 y 100. Creando un ciclo while junto con otros condicionales,  de tal forma que cuando el número sea menor o mayor al numero generado aleatoriamente te vaya dando pistas para conocer el número aleatorio.**

 Hint: import random, random.randint()



````python
# METODO DE NÚMERO PAR

# Definimos lo que vuelve "par" a un numero junto a sus condicionales
def par(numero):
  if numero % 2 == 0:
    return True
  else:
    return False

# El usuario ingresará cualquier número
numero = int(input("Ingrese un número: "))

# Se evalua el input para ver si cumple con la definición y se muestra un resultado
if par(numero):
  print("El número es par.")
else:
  print("El número no es par.")
````

````python
# CALCULAR ÁREA CON ECUACIÓN ESPECIFICA

# Importamos la libreria de matematicas
import math

# Definimos el area del poligono indicando ambos parametros y con la operación a realizar
def area_poligono(n, s):
  return n * s / (4 * math.tan(math.pi / n))

# El usuario ingresará el número de lados y la longitud de cada lado
n = int(input("Ingrese el número de lados del polígono: "))
s = float(input("Ingrese la longitud de cada lado del polígono: "))

# Se calcula el área del polígono
area = area_poligono(n, s)

# Se muestra el resultado
print("El área del polígono es:", area)
````

````python
# VERIFICAR NACIONALIDAD DE NUMERO

# Definimos la condicional para restringir los primeros 2 digitos a "52"
def Mex(telefono):
  return telefono[:2] == "52"

# El usuario ingresará el numero a verificar
telefono = input("Pasame tu numero ;) :")

# Se verificará la condicional e imprimirá el resultado
if Mex(telefono):
  print("Te llamo babe ;D eres de Mexico")
else:
  print("Eres extranjera? esto no funcionará :(")
````

````python
# CALCULO SIN USAR "m=c(1+i)**n"

# Importamos la libreria de matematicas
import math

# Definimos el monto junto a sus parametros y la operación que se realizará.
# (Pow se utiliza para exponenciar los argumentos que necesitamos para no usar "**" comodamente :))
def monto(c, i, n):
  return c * math.pow((1 + i), n)

# El usuario ingresará los datos de los argumentos
# Como dos de ellos pueden contener decimales utilizamos float
c = float(input("Ingrese el capital inicial: "))
i = float(input("Ingrese la tasa de interés: "))
n = int(input("Ingrese el número de periodos: "))

# Se calcula el monto
m = monto(c, i, n)

# Se muestra el resultado
print("El monto es:", m)
````

````python
# JUEGO DE ADIVINAR UN NUMERO

# Importamos la libreria de generación aleatoria
import random

# Definimos el numero a adivinar y el intervalo con el que trabajamos
numero_secreto = random.randint(1,100)
uwu = 0

# Definimos el bucle y el input del usuario
while uwu := numero_secreto:
  uwu = int(input("Adivina el numero entre 1 y 100: "))

# Si el input es pequeño, se indica que debe incrementar
  if uwu < numero_secreto:
    print("El numero es mayor, intenta de nuevo.")
# Si el input es grande, se indica que debe decrecer
  elif uwu > numero_secreto:
    print("El numero es menor, intenta de nuevo.")
# Si el input es igual al numero secreto, ganaste. Pero volveras a jugar indefinidamente >:)
  else:
    print("Acertaste! Ahora hazlo de nuevo :)")
````
